```python
import pandas as pd
import os
```

```python
# Define o caminho do diretório
diretorio = 'recursos'
# Lista todos os arquivos e diretórios no diretório especificado
arquivos = os.listdir(diretorio)

# Filtra para listar apenas arquivos, ignorando subdiretórios
arquivos = [arquivo for arquivo in arquivos if os.path.isfile(os.path.join(diretorio, arquivo))]

# Cria um dataframe com i nome dos arquivos.
dfArquivos = pd.DataFrame(arquivos, columns=["Arquivos"])

print(dfArquivos)
```
